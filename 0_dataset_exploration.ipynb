# Dataset description
This notebook is meant to explore <a href="http://yaroslavvb.blogspot.com/2011/09/notmnist-dataset.html">notMNIST</a> dataset with python. notMNIST dataset is similar to <a href="http://yann.lecun.com/exdb/mnist/">MNIST</a> dataset. There are 10 classes, with letters A-J taken from different fonts. Here are some examples of letter "A":
![](images/10notMNIST.png)
The dataset looks a little more like real data than MNIST: it's a harder task, and the data is a lot less 'clean' than MNIST.
I have already performed simple curation on the data. 

```python
# These are all the modules we'll be using later. Make sure you can import them
# before proceeding further.
import numpy as np
from six.moves import cPickle as pickle
import matplotlib.pylab as plt
import matplotlib.cm as cm
%pylab inline
```

# Data loading
we load first the dataset. this latter is split on tree subset: training set, validation set and test set. The datset contains images of 28 x 28 pixels.

```python
pickle_file = 'data/notMNIST.pickle'

with open(pickle_file, 'rb') as f:
  save = pickle.load(f)
  train_dataset = save['train_dataset']
  train_labels = save['train_labels']
  valid_dataset = save['valid_dataset']
  valid_labels = save['valid_labels']
  test_dataset = save['test_dataset']
  test_labels = save['test_labels']
  del save  # hint to help gc free up memory
  print('Training set', train_dataset.shape, train_labels.shape)
  print('Validation set', valid_dataset.shape, valid_labels.shape)
  print('Test set', test_dataset.shape, test_labels.shape)
```

```python
# look at the first 10 items of train_labels
# train_labels is a list of labels, each labels corresponds to one letter
# for example the label 0 corrsponds to the letter A, the label 1 to the letter B and so on.
print(train_labels[:10])
```

```python
# look at the first two items of train_dataset
# train_labels is a list of matrices. Each matrice is  
print(train_dataset[:2, :, :])
```

# Data visualization

```python
plt.rcParams['figure.figsize'] = 10, 10
# letter A corresponds to label 0
A_index = np.where(train_labels == 0)[0]
# we only visualize the 10 first images 
for (i, idx) in enumerate(A_index[:9], start=1):
    plt.subplot(3, 3, i)
    plt.imshow(train_dataset[idx, :, :], cmap=cm.Greys_r)
```
